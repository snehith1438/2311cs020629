```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name) # Use the file_name variable which contains the uploaded file's name as a string.

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number']) # This line selects columns with numerical data types

# Calculate the correlation matrix on the numerical columns
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Price~Mfg_Year+Central_Lock+Powered_Windows',data=df).fit()
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
import pandas as pd
df = pd.read_csv("Day_13_Pharma_data.csv")
df.head()
```

```python
missing_values = df.isnull().sum()
duplicates = df.duplicated().sum()
df_cleaned = df.dropna()
print("Missing values per column:\n", missing_values)
print("\nNumber of duplicate rows:", duplicates)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df_cleaned, estimator=sum, errorbar=None)
plt.title('Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df_cleaned)
plt.title('Relationship between Marketing Spend and Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=df_cleaned)
plt.title('Drug Effectiveness Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.show()
```

```python
correlation = df_cleaned[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap between Sales, Marketing Spend, and Effectiveness')
plt.show()
```

```python
!pip install scikit-learn  # Installs the scikit-learn library if not already installed
import pandas as pd
from google.colab import files
from sklearn.cluster import KMeans # Imports the KMeans class from sklearn.cluster
import matplotlib.pyplot as plt # Imports the pyplot module for plotting, aliased as plt

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

km = KMeans(n_clusters=3)
y_predicted = km.fit(df[['Age','Income($)']])

df['cluster']=y_predicted

df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green') # Now plt is defined and can be used
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
```
